### (4) X,y 컬럼 선택 및 train, test data로 분리하기

- datetime 형태는 Linear Regression에 사용할 수 없으므로 object형으로 바꾸려고 코드를 작성했는데 확인해보니 datetime type은 'datetime' 컬럼 하나밖에 없으므로 indexing을 이용해서 해당 컬럼만 제외해도 됐었다.

```python
# df의 datetime type 데이터 조회

datetime_columns = []

for column in df.columns:
    if df[column].dtype == 'datetime64[ns]':
        datetime_columns.append(column)

datetime_columns
```

```python
# datetime 타입을 object형으로 변경
df[datetime_columns] = df[datetime_columns].astype('object')

df.info()
```

```python
# train/test 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:11], df.iloc[:,11], test_size=0.2, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### (5)  Linear  Regression 모델 학습하기

```python
# df의 datetime type 데이터 조회

datetime_columns = []

for column in df.columns:
    if df[column].dtype == 'datetime64[ns]':
        datetime_columns.append(column)

datetime_columns
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

### (6) 학습된 모델로 X_test에 대한 예측값, 손실함수 값 계산하기 

```python
pred = model.predcit(X_test)
```
